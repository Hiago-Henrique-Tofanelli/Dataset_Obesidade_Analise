```python
import pandas as pd
```

```python
obesityDT = pd.read_csv('Conditions_Contributing_to_COVID-19_Deaths__by_State_and_Age__Provisional_2020-2023.csv')
obesityDT
```

```python
# Filtrando o DataFrame para incluir apenas as entradas com 'State' igual a 'United States'
obesityDT_us = obesityDT[obesityDT['State'] == 'United States']

# Agora, 'obesityDT_us' contém apenas as entradas para o estado 'United States'
```

```python
obesityDT_us
```

```python
# Filtrando o DataFrame para incluir apenas as entradas com 'State' igual a 'United States' e 'Condition' igual a 'obesity'
obesityDT_filtered = obesityDT[(obesityDT['State'] == 'United States') & (obesityDT['Condition'] == 'Obesity')]

# Agora, 'obesityDT_filtered' contém apenas as entradas para o estado 'United States' com 'Condition' igual a 'obesity'
```

```python
obesityDT_filtered
```

```python
# Removendo as colunas "Year", "Month" e "ICD10_codes" do DataFrame
obesityDT_filtered = obesityDT.drop(columns=['Year', 'Month', 'ICD10_codes'])

# Agora, 'obesityDT_filtered' contém as demais colunas, exceto "Year", "Month" e "ICD10_codes"
```

```python
obesityDT_filtered
```

```python
obesityDT_filtered = obesityDT_filtered[(obesityDT_filtered['Age Group'] != 'All Ages') & (obesityDT_filtered['Age Group'] != 'Not stated')]

obesityDT_filtered = obesityDT_filtered[(obesityDT_filtered['State'] == 'United States') & (obesityDT_filtered['Condition'] == 'Obesity')]

obesityDT_filtered = obesityDT_filtered[(obesityDT_filtered['Group'] == 'By Total')]

obesityDT_filtered
```

```python
# Substituindo a numeração das linhas pelos valores de 1 a 7
obesityDT_filtered.index = range(1, 9)

obesityDT_filtered
```

```python
# Removendo a coluna "Flag"
# obesityDT_filtered = obesityDT_filtered.drop(columns=['Flag'])

# obesityDT_filtered = obesityDT_filtered.drop(columns=['Number of Mentions'])

obesityDT_filtered = obesityDT_filtered.drop(columns=['Condition'])

obesityDT_filtered
```

```python
# Calculando a soma do número de mortos na coluna "COVID-19 Deaths"
soma_mortos = obesityDT_filtered['COVID-19 Deaths'].sum()

# Exibindo o resultado da soma
print(f'O número total de mortos é: {soma_mortos}')

```
